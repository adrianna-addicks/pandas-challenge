sc## School Summary

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()







#Total budget
total_budget= school_data["budget"].sum()


#Student Count
student_count=student_data['Student ID'].count()


#Total number of schools
school_count=school_data['School ID'].count()

#Avg math score
avg_math_score=student_data['math_score'].mean()


#Avg reading score
avg_read_score=student_data['reading_score'].mean()

#Passing math score over 70
passing_math=school_data_complete[school_data_complete['math_score']>=70].count()['Student ID']
#Passing percentage
passing_math_percent=passing_math/student_count*100

#Passing reading score over 70
passing_reading=school_data_complete[school_data_complete['reading_score']>=70].count()['Student ID']

#Passing percentage
passing_reading_percent=passing_reading/student_count*100
#Overall passing
overall_passing=school_data_complete[(school_data_complete['reading_score']>=70)&(school_data_complete['math_score']>=70)].count()['Student ID']
 #Overall passing percentage                                   
overall_passing_percent=overall_passing/student_count*100
#Overview summary table
district_summary=pd.DataFrame({"Total Schools": [school_count],
                              "Total Students": [student_count],
                              "Total Budget":   [total_budget],
                              "Average Math Score": [avg_math_score],
                              "Average Reading Score": [avg_read_score],
                              "% Passing Math": [passing_math_percent],
                              "% Passing Reading": [passing_reading_percent],
                              "% Overall Passing": overall_passing_percent})
#formatting
district_summary["Total Students"]=district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)


#school_data_complete.columns
#Index(['Student ID', 'student_name', 'gender', 'grade', 'school_name',
       #'reading_score', 'math_score', 'School ID', 'type', 'size', 'budget'],
#School name
school_types = school_data.set_index(["school_name"])["type"]
#total students
per_school_counts = school_data_complete["school_name"].value_counts()
#school budget
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
#Per student budget
per_school_capita = per_school_budget / per_school_counts
#avg math score
per_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
#avg reading score
per_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
#passing math info by schools
school_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
#passing reading scores by schools
school_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
#combined passing scores
passing_math_and_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)
                                              & (school_data_complete["math_score"] >= 70)]
#using math/reading scores grouped by school to determine percentage
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
# turning overall passing rate into percentage
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100

# Convert to DataFrame
per_school_summary = pd.DataFrame({"School Type": school_types,
                                  "Total Students": per_school_counts,
                                  "Total School Budget": per_school_budget,
                                  "Per Student Budget": per_school_capita,
                                  "Average Math Score": per_school_math,
                                  "Average Reading Score": per_school_reading,
                                  "% Passing Math": per_school_passing_math,
                                  "% Passing Reading": per_school_passing_reading,
                                  "% Overall Passing": overall_passing_rate})
#formatting
per_school_summary["Total School Budget"]=per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"]=per_school_summary["Per Student Budget"].map("${:,.2f}".format)

#top performing schools
sorting_schools=per_school_summary.sort_values("% Overall Passing", ascending=False)

sorting_schools[0:5]
#bottom performing schools
bottom_sorting_schools=per_school_summary.sort_values("% Overall Passing")

bottom_sorting_schools[0:5]

```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()

#Index(['Student ID', 'student_name', 'gender', 'grade', 'school_name',
       #'reading_score', 'math_score', 'School ID', 'type', 'size', 'budget'],
      #dtype='object')


school_data_complete.head()



#grade_sort_column=school_data_complete[["grade","school_name", "reading_score"]]
#grade_sort_column.head()

#columns=["grade",
 #       "school_name",
 #       "reading_score"]

# Create data series of scores by grades using conditionals
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]


# Group by each school for math avg score
math_ninth_graders_scores = ninth_graders.groupby(["school_name"]).mean()["math_score"]
math_tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
math_eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
math_twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()["math_score"]



# Math Scores by grade summary
math_scores_by_grade= pd.DataFrame({"9th":math_ninth_graders_scores,
                                    "10th":math_tenth_graders_scores,
                                     "11th":math_eleventh_graders_scores,
                                   "12th":math_twelfth_graders_scores})
            
 #output          
math_scores_by_grade.head(15)

# Group each by school name for reading avg score


reading_ninth_graders_scores = ninth_graders.groupby(["school_name"]).mean()["reading_score"]
reading_tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()["reading_score"]
reading_eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()["reading_score"]
reading_twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()["reading_score"]


#Reading Scores by grade summary
reading_scores_by_grade= pd.DataFrame({"9th":reading_ninth_graders_scores,
                                    "10th":reading_tenth_graders_scores,
                                     "11th":reading_eleventh_graders_scores,
                                   "12th":reading_twelfth_graders_scores})


#output
reading_scores_by_grade.head(15)
```
